# Data Visualization with Seaborn

Data visualization is an important aspect of data science. It allows data to be understood more easily. 

No longer are the days where charts, graphs and Microsoft Excel spreadsheet dominates. Nowadays, there are a variety of sophisticated ways to illustrate data. Examples include heatmaps, infographics, geographic maps, and more.

This project will explore data visualization using the Dota2 hero datasets. Dota2 is a popular computer game developed by Valve Corporation. The datasets contains numerical features like Health and Intelligence, as well as categorical features such as Primary Stat and Attack Type.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

# Data Preprocessing
       
Data preprocessing transforms raw data into a format that is easier to work with and understand. 
    
In real life, data are often incomplete, inconsistent, lacking in certain trends, or contains 
unneccesary information. 


```python
df = pd.read_csv("hero_stats.csv")
df.head()
```

Notice that there are 45 columns and 115 rows, but the rows and columns in between are truncated. Pandas has the option to display all the rows and columns.

```python
pd.options.display.max_columns = None
pd.options.display.max_rows = None
df.head()
```

Features such as bonus damage, bonus damage reduction, evasion, bash chance, crit chance, miss chance and lifesteal are not intrinsic to the heros. They are only acquired at later stages in the game. These features can be dropped.

```python
df.drop(['Bonus Damage','Bonus Damage Reduction', 'Evasion', 'Bash Chance', 'Crit Chance','Miss Chance','Lifesteal'], axis = 1, inplace=True)
df.head()
```

# Relational Plot

Relational plot is one of the most basic ways to visualize the relationship between two variables.

Relational plot has the advantage of showing linear and non-linear relationships between variables. Relational plots are also easy to interpret.

```python
sns.relplot(x="Strength", y="Strength Gain", hue="Primary Stat", data=df);
```

The graph above shows a positive correlation between strength and strength gain. This makes sense, as Strength Heroes 
on average enjoys a higher strength gain.

Refer to https://dota2.gamepedia.com/Strength for more info on strength.

We can see draw regression lines to better see relationship between Strength and Strength Gain of each Hero, using Seaborn's lmplot.

```python
sns.lmplot(x="Strength", y="Strength Gain", hue="Primary Stat", data=df, ci=None);
```

However, the relationship is a weak positive correlation. Having a higher base Strength does not imply having a higher strength gain. Some heroes start with high base strength, but does not scale as well as the game goes on.

We can also see a multivariate relationship between features. For example, lets look at the relationship between agility, armor and physical damage reduction.

We will use mplot3d to plot 3d plots.  

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D

plot = plt.figure().gca(projection='3d')
plot.scatter(df['Agility'], df['Armor'], df['%Physical Damage Reduction'])
plot.set_xlabel('Agility')
plot.set_ylabel('Armor')
plot.set_zlabel('%Physical Damage Reduction')
plt.show()
```

Unfortunately, IPython does support very interactive 3d plots.

# Box Plots

Box plots are simple visualizations of data through their quartiles.

The whiskers represents the min and max values, the box encompasses the interquartile range, and the line in the middle of the box represents the median.

Let us look at the whisker plot of Intelligence heroes. We will look at the Intelligence, Intelligence Gain, 
Mana and Mana Regeneration.

```python
df_agi = df[df["Primary Stat"] == 'AGI']
df_agi = df[['Agility', 'Armor', '%Physical Damage Reduction']]
sns.boxplot(data = df_agi)
```

There are datapoints that lies outside of the box plots, represented by the diamond shape. 
These outliers are determined by Seaborn using a function of the interquartile range. 

There exists a hero with zero agility. We can use Panda's row conditional to locate it.

```python
df[df['Agility'] == 0]
```

Turns out Tiny is the hero that starts with absolutely no Agility or Agility gain. Luckily, he offsets these disadvantages via items and hero skills.

# Heat Maps

Heat maps provide a graphical way to determine the correlation between features.
Correlation is a statistical technique that shows how close two variables are related to each other.



```python
#remove categorical features
df_num = df.select_dtypes(include=['float', 'int'])
df_num.head()
```

Next calculate the correlation matrix (truncated).

```python
corr = df_num.corr()
corr.head()
```

Correlation matrix shows the level of linear relationship between two variables. Positive number indicates positive relationship. Negative number indicates negative relationship.

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.heatmap(corr)
```

Unsurprisingly, there is a very strong correlation (white) between the damage attributes (Base Damage Average, Base Damage Min, Base Damage Max, Damage Average, Damage min, Damage Max).

There is a negative correlation (black) between Agility and Attack Time, since each point of Agility decreases Attack Time.

There is no correlation between Agility and Base Attack Time (red). Base Attack Time is intrinsic to each hero, and cannot be modified by any stats. Another attribute that has no correlation to other attributes also is Projectile Speed, which is intrinsic to the hero.

Heatmap is a good two dimensional way to see the relationship between multivariable datasets.

# Histograms

Histograms enable the visualization of numerical values. It can display the frequency of observations, or normalized to display the probability instead.

```python
sns.distplot(df["Intelligence"])
```

A kernel density estimate is used to approximate a normal distribution. Because the hero's base Intelligence are discrete values, a KDE is probably not a good approximation. This can be fixed as shown below. 


```python
sns.distplot(df["Intelligence"], kde = False)
```

# Bar Plots
Bar plots allows categorical data to be illustrated. Do not confuse bar plots with histograms. Histograms shows the distribution of variables, while bar plots are used to compare variables.

```python
df_cat = df[['Primary Stat', 'Attack Type']]
sns.countplot(x='Primary Stat',hue = "Attack Type", data=df_cat)
```

Strength Heroes have high health and high survivability, and are thus mostly melee Heroes.
Intelligence Heroes are squishy, therefore more suited for ranged attacks.
Agility Heroes are more split more evenly between melee and ranged attacks.

# Violin Plots

Violin plots are the combination of box plots and kernel density plot. 
It combines the advantages of both plots: displaying summary statistics as well as data distribution.
Violin plots can also display multimodal peaks, something that box plots can't accomplish.

```python
ax = sns.catplot(kind="violin", data=df_agi)
ax.fig.suptitle("Violin Plots for Agility Heroes")
```

Agility violin plot has two peaks that can be displayed.

# KDE Plots

Better know as contour plots, these are basically kernel density estimates for bivariate datasets. 

```python
sns.kdeplot(df['Movement Speed'], df['Turn Rate'], shade=True, cbar=True)
```

Heroes with turn rate that are significantly higher? 

```python
df[df["Turn Rate"] > 0.9]
```

Most heros with a turn rate of 0.5 and 0.6 would have a movement speed of between 280 and 300. Interesting find.

# Joint Distribution Plot

Joint distribution plot combines univariate and bivariate plots into one.

Let's look at the bivariate relationship between Physical Damange Reduction and Physical Effective Health.

According to Dota 2 Wiki, the Physical Effective Health is a function of both HP and Armor.

EHP = current_HP * (1 + ( 0.052 * current_armor ) / ( 0.9 + 0.048 * |current_armor|))

```python
sns.jointplot(x='%Physical Damage Reduction', y='Physical Effective Health', data=df)
```

Joint distribution plot is not limited to display scatter plot.

```python
sns.jointplot(x='%Physical Damage Reduction', y='Physical Effective Health', data=df, kind="kde")
```

```python
sns.jointplot(x='%Physical Damage Reduction', y='Physical Effective Health', data=df, kind="reg")
```

Joint probability distribution shows the probability of an observation that satisfies both variables.

Let's look at some outliers.

```python
df[df['%Physical Damage Reduction'] > 30]
```

```python
print("Armor Mean: " + str(df['Armor'].mean()))
print("EHP Mean: " + str(df['Physical Effective Health'].mean()))
ehp = df[df['Physical Effective Health'] > 900]
ehp
```

These 5 heroes have average starting Health, but their above average Armor adds a significant chunk to their Physical Effective Health (from 711 to the 900s).
